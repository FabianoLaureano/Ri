Primeiramente obter os dados contidos no arquivo csv

```python
import pandas as pd
```

```python
data = pd.read_csv("noticias_estadao.csv")
data
```

```python
dic = {}
for linha in data.iterrows():
        terms = linha[1]['conteudo'].split(' ')
        for term in terms:
            term = term.lower()
            if term in dic.keys():
                if linha[1]['idNoticia'] not in dic[term]:
                    dic[term].append(linha[1]['idNoticia'])
            else:
                dic[term] = [linha[1]['idNoticia']]
```

Algoritmos que auxiliam na busca:
    
    search: usuário passa um termo e o algoritmo retorna os documentos quem contém aquele termo;
    intersect: retorna os documentos que contém as mesmo tempo os dois termos na pesquisa;
    union: retorna os documentos que contém ou um o outro termo pesquisados.


```python
def intersect(p1, p2):
    p1.sort()
    p2.sort()
    answer = []
    i = 0
    j = 0
    while (i < len(p1) and j < len(p2)):
        if int(p1[i]) == int (p2[j]):
            answer.append(p1[i])
            i = i + 1
            j = j + 1
        else:
            if int(p1[i]) < int (p2[j]):
                i = i + 1
            else:
                j = j + 1 

    return answer

def union(p1, p2):
    p1.sort()
    p2.sort()
    answer = []
    i = 0
    j = 0

    for elemento in p1:
        if not elemento in answer:
            answer.append(elemento)
    for elemento in p2:
        if not elemento in answer:
            answer.append(elemento)

    return answer

def search(term):
    return dic[term]
```

Asserts usando a seguinte regra:
    
    tamanho da união entre A e B = tamanho de A + tamanho de B - tamanho da interseção entre A e B 

```python
assert len(union(dic['belo'], dic['horizonte'])) == len(search('belo')) + len(search('horizonte')) - len(intersect(dic['belo'], dic['horizonte']))

assert len(union(dic['presidenciáveis'], dic['corruptos'])) == len(search('presidenciáveis')) + len(search('corruptos')) - len(intersect(dic['presidenciáveis'], dic['corruptos']))

assert len(union(dic['debate'], dic['presidencial'])) == len(search('debate')) + len(search('presidencial')) - len(intersect(dic['debate'], dic['presidencial']))

assert len(intersect(dic['presidenciáveis'], dic['corruptos'])) == 0
```
